BeautifulSoup使用

```python
from bs4 import BeautifulSoup
import requests

url = "http://www.baidu.com"
html = requests.get(url).text
# 使用BeautifulSoup(html,"html.parser")将html解析后的数据存入sp
sp = BeautifulSoup(html,"html.parser")

# 输出该网页的标题
print(sp.title)
# 除去所有的HTML标签，把网页变成字符串返回
print(sp.text)
# 返回以CSS选择器作为运算结果的所有内容，主要操作对象为id和class
print(sp.select("#f60"))

#返回第一个符合条件的内容
print(sp.find('button'))

# 提取网页中的链接标签
links = sp.find_all('a')
for i in range(1,10):
#     输出所有链接标签
    print(links[i])
#     输出该链接标签内容
    print(links[i].contents)
#     输出该链接标签href属性值
    print(links[i].get('href'))
    
```

```python
from bs4 import BeautifulSoup
import requests
import sys

kw = '西安建筑科技大学'
red_tag = 'i2937740099'

#获取将要访问的所有网址
def init_addr():
	address = []
	i = 0
	while i < 10000:
		url2 = 'http://tieba.baidu.com/f?kw='+kw+'&ie=utf-8&pn='+str(i)
		address.append(url2)
		i += 50
	return address

#获取该网址所在html 
def get_html_data(url):
    html = requests.get(url).text
    sp = BeautifulSoup(html,'html.parser')
    return sp

#获取该网页中所有帖子的网址
def find_href(url):
    hrefs=[]
    sp = get_html_data(url)
    all_links = sp.find_all('a')
    print(all_links)
    # 检查该网页中所有帖子的网址
    for link in all_links:
        href = link.get('href')
        if href != None and href.startswith('/p/'):
            href = 'http://tieba.baidu.com' + href
            hrefs.append(href)
            print(href)
            file = open('所有帖子的网址.txt','a',encoding='utf-8')
            file.write(href + '\n')
    return hrefs


if __name__ == '__main__':
#获取将要访问的所有网址,并保存至文件
	address = init_addr()
	file = open('将要访问的网址.txt','w',encoding='utf-8')
	for addr in address:
		file.write(addr + '\n')
	#逐个遍历已存在网页
	tiezi_links = []
	for addr in address:
		tiezi_links.append(find_href(addr))
	
	
	
	
	
	
	
```
