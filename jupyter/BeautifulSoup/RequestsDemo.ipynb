提取网页内容

```python
import requests

url = "http://tieba.baidu.com/f?kw=%E8%A5%BF%E5%AE%89%E5%BB%BA%E7%AD%91%E7%A7%91%E6%8A%80%E5%A4%A7%E5%AD%A6&fr=index&red_tag=i2937740099"
# 使用requests.get(url)提取此网页的内容
# text.splitlines() 将文本按行分割为列表
html = requests.get(url).text.splitlines()
# 遍历列表
for i in range(0,1500):
    print(html[i])
    
    
```
